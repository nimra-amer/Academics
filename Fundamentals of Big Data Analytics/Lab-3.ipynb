```python
#TASK-A

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Mall_Customers.csv")
x= data['Annual Income (k$)']
y = data["Spending Score (1-100)"]

plt.scatter(x,y)
```

```python
from sklearn.cluster import KMeans
data = list(zip(x, y))
inertias = []

for i in range(1,15):
 kmeans = KMeans(n_clusters=i)
 kmeans.fit(data)
 inertias.append(kmeans.inertia_)

plt.plot(range(1,15), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

```python
#TASK_B
from numpy import random
import math
random_x_1 = random.choice(x)
random_y_1 = random.choice(y)

centroid1 = (random_x_1,random_y_1)
print("Centroid 1 : " , centroid1)

random_x_2 = random.choice(x)
random_y_2 = random.choice(y)

centroid2 = (random_x_2,random_y_2)
print("Centroid 2 : " , centroid2)
```

```python
import numpy as np
#Calculating Euclidean Distance
distance_1 = [None]*200
distance_2 = [None]*200
for i in range(0,200):
  distance_1[i] = ((random_x_1 - x[i])**2 + (random_y_1 - y[i])**2)**(1/2)
  distance_2[i] = ((random_x_2 - x[i])**2 + (random_y_2 - y[i])**2)**(1/2)

#Comparing The Distances To Make Clusters
cluster_1 = [None]*80
cluster_2 = [None]*120
j = 0
k=0
for i in range(0,200):
  if distance_1[i] <= distance_2[i]:
    cluster_1[j] = i
    j = j + 1
  else:
    cluster_2[k] = i
    k = k+1

print("Indexes for First Cluster : " , cluster_1)
print("Indexes for Second Cluster : " , cluster_2)

```

```python
#Calculating mean for new centroids
sum_x = 0
sum_y = 0
sum_x_2 = 0
sum_y_2 =0

for i in range(0,80):
  sum_x = sum_x + x[cluster_1[i]]
  sum_y = sum_y + y[cluster_1[i]]
  sum_x_2 = sum_x_2 + x[cluster_2[i]]
  sum_y_2 = sum_y_2 + y[cluster_2[i]]

new_x = sum_x/80
new_y = sum_y/80

new_x_2 = sum_x_2/120
new_y_2 = sum_y_2/120

new_centroid_1 = (new_x , new_y)
print("First New_Centroid : " , new_centroid_1)

new_centroid_2 = (new_x_2 , new_y_2)
print("Second New_Centroid : " , new_centroid_2)
```

```python
#Second  Iteration
new_distance_1 = [None]*200
new_distance_2 = [None]*200
for i in range(0,200):
  new_distance_1[i] = ((new_x - x[i])**2 + (new_y - y[i])**2)**(1/2)
  new_distance_2[i] = ((new_x_2 - x[i])**2 + (new_y_2 - y[i])**2)**(1/2)

#Comparing The Distances To Make Clusters
cluster_1_new = [None]*113
cluster_2_new = [None]*87
j = 0
k=0
for i in range(0,200):
  if new_distance_1[i] <= new_distance_2[i]:
    cluster_1_new[j] = i
    j = j + 1
  else:
    cluster_2_new[k] = i
    k = k+1

print("Indexes for First Cluster : " , cluster_1_new)
print("Indexes for Second Cluster : " , cluster_2_new)
```

```python
#Calculating mean for new centroids
sum_x_2 = 0
sum_y_2 = 0
sum_x_4 = 0
sum_y_4 =0

for i in range(0,80):
  sum_x_2 = sum_x_2 + x[cluster_1_new[i]]
  sum_y_2 = sum_y_2 + y[cluster_1_new[i]]
  sum_x_4 = sum_x_4 + x[cluster_2_new[i]]
  sum_y_4 = sum_y_4 + y[cluster_2_new[i]]

new_x_2 = sum_x_2/113
new_y_2 = sum_y_2/113

new_x_4 = sum_x_4/87
new_y_4 = sum_y_4/87

new_centroid_3 = (new_x_2 , new_y_2)
print("First New_Centroid : " , new_centroid_3)

new_centroid_4 = (new_x_4 , new_y_4)
print("Second New_Centroid : " , new_centroid_4)
```

```python
#plotting k-means
plot_x_1 = x[cluster_1_new]
plot_y_1 = y[cluster_1_new]

plot_x_2 = x[cluster_2_new]
plot_y_2 = y[cluster_2_new]

plt.scatter(plot_x_1 , plot_y_1)
plt.scatter(plot_x_2 , plot_y_2)
```
