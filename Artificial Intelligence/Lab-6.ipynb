```python
# Python3 implementation of the
from random import randint

N = 8

# A utility function that configures the 2D array "board" and array "state" randomly to provide a starting point for the algorithm.
def configureRandomly(board, state):

	# Iterating through the column indices
	for i in range(N):

		# Getting a random row index
		state[i] = randint(0, 100000) % N

		# Placing a queen on the obtained place in chessboard.
		board[state[i]][i] = 1

# A utility function that prints the 2D array "board".
def printBoard(board):

	for i in range(N):
		print(*board[i])

# A utility function that prints the array "state".
def printState( state):
	print(*state)

# A utility function that compares two arrays, state1 and state2 and returns True if equal and False otherwise.
def compareStates(state1, state2):


	for i in range(N):
		if (state1[i] != state2[i]):
			return False

	return True

# A utility function that fills the 2D array "board" with values "value"
def fill(board, value):

	for i in range(N):
		for j in range(N):
			board[i][j] = value
```

```python
def calculateObjective(board, state):
    N = len(board)
    attacking = 0

    for i in range(N):
        if i >= len(state):
            break

        row = state[i]
        col = i - 1
        while col >= 0:
            if col < N and row < N:  # Check bounds
                if board[row][col] == 1:
                    attacking += 1
                    break
            col -= 1

        row = state[i]
        col = i + 1
        while col < N:
            if col >= 0 and row < N:  # Check bounds
                if board[row][col] == 1:
                    attacking += 1
                    break
            col += 1

        row = state[i] - 1
        col = i - 1
        while col >= 0 and row >= 0:
            if col < N and row < N:  # Check bounds
                if board[row][col] == 1:
                    attacking += 1
                    break
            col -= 1
            row -= 1

        row = state[i] + 1
        col = i + 1
        while col < N and row < N:
            if col >= 0 and row >= 0:  # Check bounds
                if board[row][col] == 1:
                    attacking += 1
                    break
            col += 1
            row += 1

        row = state[i] + 1
        col = i - 1
        while col >= 0 and row < N:
            if col < N and row >= 0:  # Check bounds
                if board[row][col] == 1:
                    attacking += 1
                    break
            col -= 1
            row += 1

        row = state[i] - 1
        col = i + 1
        while col < N and row >= 0:
            if col >= 0 and row < N:  # Check bounds
                if board[row][col] == 1:
                    attacking += 1
                    break
            col += 1
            row -= 1

    return attacking // 2  # Returning integer division
```
