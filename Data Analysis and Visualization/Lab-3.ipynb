### **5. Encode categorical features**



1.   Label Encoding
2.   OneHot Encoding
3.   Target Encoding
4.   Binary Transformation


```python
#Label Encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

data['StoreType'] = label_encoder.fit_transform(data['StoreType'])
data['Assortment'] = label_encoder.fit_transform(data['Assortment'])
```

```python
reshape_ = data['StoreType'].values.reshape(-1,1)
reshape_1 = data['Assortment'].values.reshape(-1,1)

#OneHotEncoding

from sklearn.preprocessing import OneHotEncoder
Onehot_encoder = OneHotEncoder(sparse=False)
Store = Onehot_encoder.fit_transform(reshape_)
Assor = Onehot_encoder.fit_transform(reshape_1)

data['StoreType'] = Store.tolist()
data['Assortment'] = Assor.tolist()
```

```python
#Target Encoding

from sklearn.preprocessing import TargetEncoder

Targetencoder = TargetEncoder()
data['StoreType'] = Targetencoder.fit_transform(data['StoreType'])
data['Assortment'] = Targetencoder.fit_transform(data['Assortment'])
```

```python
#Binary Transformation

import category_encoders as ce
data['StoreType'] = ce.BinaryEncoder().fit_transform(data['StoreType'])
data['Assortment'] = ce.BinaryEncoder().fit_transform(data['Assortment'])
```

What is the average daily sales for each store in the train_store dataframe?

```python
average_sales = data.groupby('Store')['Sales'].mean().reset_index()
average_sales.rename(columns={'Sales': 'Average'}, inplace=True)

print(average_sales)
```

How many days of data are included in the train_store dataframe, and what is the range of dates?

```python
data_ = pd.merge(data_train , data_store) #merging again because we have dropped the date column for our pre processing

data_['Date'] = pd.to_datetime(data_['Date'])

# Find the minimum and maximum dates
min_date = data_['Date'].min()
max_date = data_['Date'].max()

num_days = (max_date - min_date).days + 1

print("Number Of Days :", num_days)
print("Range :", min_date, "to", max_date)
```

What is the average sales per customer for each store in the train_store dataframe?

```python
average_sales = data_.groupby('Store').apply(lambda group: (group['Sales'] / group['Customers']).mean())
print("Average Sales Per Customer : \n" , average_sales)
```

How many days of the week does each store in the train_store dataframe open and close?

```python
open_days = data_[data['Open'] == 1].groupby('Store')['DayOfWeek'].unique()
print("Open Days : \n")
print(open_days)

print("Close Days : \n")
close_days = data_[data['Open'] == 0].groupby('Store')['DayOfWeek'].unique()
print(close_days)
```

What is the correlation between the Sales and Customers columns in the train_store dataframe?
