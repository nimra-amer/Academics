TASK 1 (B): Perform Tokenization Technique on Raw Text File

```python
from nltk.tokenize import word_tokenize

# Tokenizes the sentences meaning split the sentences into words seperated by the "white sapce".
# input - List of sentences
# returns a list of lists of each sentence being tokenized.

def tokenize_sentence(lines):
    ### Write your Code here
    tk = WhitespaceTokenizer()
    token = []
    for line in lines:
        tokens = tk.tokenize(line)
        token.append(tokens)
    
    return token
```

```python
# remove punctuations -print(string.punctuation) ---- !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~ ----
# remove empty strings.
# lower case all the words
# add <s> at the beginning and </s> at the end of every sentence in the corpus.
# input - list of lists of words obtained from "tokenize_sentence" function.
# returns - list of lists
def prep_data(lines):
    ### Write your Code here # remove punctuations
        
        array = []
        for line in lines:
                  if line:    # removes empty strings
                    low = ''.join([c.lower() if 'A' <= c <= 'Z' else c for c in line]) # lower case
                    low = low + ' </s>'  # Append </s> at the end of each sentence in the corpus
                    low = '<s>' + low # Append <s> at the beginning of each sentence in the corpus
                    array.append(low)  


        print("No of sentences in Corpus: "+str(len(lines)))
        print("\n")
        return array
```

```python
# Counts the no. of times a word repeats (frequency of each word) in the corpus.
# input - list of lists of words obtained from "prep_data"
# returns - a dictionary defined as {word:frequency} for words of the corpus including <s> and </s>.
from collections import Counter
def freq_of_unique_words(lines):
   ### Write your Code here
   words = [word for line in lines for word in line.split()]
   word_frequencies = Counter(words)

   return word_frequencies
```

```python
'''
    Bayes Theorem form
    P(y|X) = P(X|y) * P(y) / P(X)
    '''
def calc_prior(features, target):
        '''
        prior probability P(y)
        calculate prior probabilities
        '''
        ### Write your code here
        data_ = pd.concat([features, target], axis=1)
    
        prior_prob = {}
    
        for f, g in data_.groupby(list(features.columns)):
         class_counts = g.iloc[:, -1].value_counts()
         total_instances = len(g)
         prior_prob[f] = class_counts / total_instances
         
        return prior_prob

def calc_statistics(features, target):
        '''
        calculate mean, variance for each column and convert to numpy array
        '''
         ### Write your code here
        mean_target = target.mean()
        mean_features = features.mean()

        variance_target = target.var()
        variance_features = features.var() 

        mean = np.array([str(mean_target)] + mean_features.to_list())
        variance = np.array([str(variance_target)] + variance_features.to_list())
    
          
        return mean, variance

def gaussian_density(x):
        '''
        calculate probability from gaussian density function (normally distributed)
        we will assume that probability of specific target value given specific class is normally distributed

        probability density function derived from wikipedia:
        (1/√2pi*σ) * exp((-1/2)*((x-μ)^2)/(2*σ²)), where μ is mean, σ² is variance, σ is quare root of variance (standard deviation)
        '''
        ### Write your code here
        std_deviation = 1  # Standard deviation for a standard normal distribution.
        exponent = -0.5 * (x ** 2)
        prob = (1 / (np.sqrt(2 * np.pi) * std_deviation)) * np.exp(exponent)
    
        return prob


def calc_posterior(x , features , target):
         posteriors = []
    
        # calculate posterior probability for each class
         ### Write your code here
         class_means , class_variances = [calc_statistics(features, target)]
         posterior_probabilities = []
         for i in range(2):  # 
          posterior = gaussian_density(x, class_means[i], class_variances[i])
          posterior_probabilities.append(posterior)
    
         if posterior_probabilities[0] >= posterior_probabilities[1]:
            highest_posterior_class = 0 
         else: 
             highest_posterior_class = 1

         return highest_posterior_class, posterior_probabilities
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
column_names = [
    "Age", "Work", "Fnlwgt", "Education", "Education_num", "Marital_status",
    "Occupation", "Relationship", "Race", "Gender", "Capital_gain", "Capital",
    "hours_per_week", "Country", "Income"
]
data = pd.read_csv("adult_data - adult_data.csv", header=None, names=column_names)

X = data.drop('Income' , axis = 1)
Y = data['Income']
X = X.dropna()
Y = Y.dropna()


categorical_columns = ["Work", "Education", "Marital_status", "Occupation", "Relationship", "Race", "Gender", "Country"]



X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train = X_train.dropna()
y_train = y_train.dropna()
X_train = pd.get_dummies(X_train, columns=categorical_columns)

X_train['Work'] = le.fit_transform(X_train['Work'])
X_train['Education'] = le.fit_transform(X_train['Education'])
X_train['Martial_status'] = le.fit_transform(X_train['Martial_status'])
X_train['Occupation'] = le.fit_transform(X_train['Occupation'])
X_train['Relationship'] = le.fit_transform(X_train['Relationship'])
X_train['Race'] = le.fit_transform(X_train['Race'])
X_train['Gender'] = le.fit_transform(X_train['Gender'])
X_train['Country'] = le.fit_transform(X_train['Country'])

nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

y_pred = nb_classifier.predict(X_test)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f_score = f1_score(y_test, y_pred)

print("Accuracy: ", acc*100 , "%")
print("Precision: " , prec*100 , "%")
print("Recall: " , recall*100 , "%")
print("F1 Score: ",f_score)
```
